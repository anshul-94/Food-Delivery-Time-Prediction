```python
missing_summary = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing Percentage": (
        df.isnull().sum() / len(df)
    ) * 100
})

missing_summary = missing_summary[
    missing_summary["Missing Count"] > 0
]

missing_summary.sort_values(
    by="Missing Percentage",
    ascending=False
)
```
